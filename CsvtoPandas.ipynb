```python
targetcsv = r"C:\Users\juncy\bus_data/"
output_file = r"C:\Users\juncy\output\NumofBusRider.csv"
csv_files = os.listdir(targetcsv)
#print(csv_files)

file_list = glob.glob(targetcsv+'*.csv')
#print(file_list)

with open(output_file, 'w') as f:
    for i, file in enumerate (file_list):
        if i==0:
            with open(file,'r', encoding='UTF-8') as f2:
                while True:
                    line=f2.readline()
                    if not line:
                        break
                    f.write(line)
            print(file.split('\\')[-1])
        else:
            with open(file,'r', encoding='UTF-8') as f2:
                LineNo=0
                while True:
                    line=f2.readline()
                    if LineNo!=0: #2번째 파일부터는 첫번째 줄(헤더)제외
                        f.write(line)
                    if not line:
                        break
                    LineNo+=1
            print(file.split('\\')[-1])
```
