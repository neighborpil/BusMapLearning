```python
import csv

def write(text):
  with open('bus_stops.csv', 'a', encoding='UTF8') as file:
    file.write(text)
    file.write('\n')
```

```python
bus_numbers = []
for i in range(10):
    search_input = driver.find_element('xpath', '//*[@id="bline_sch"]/dl/dd/input')
    search_input.click()
    search_input.clear()
    search_input.send_keys(str(i))
    time.sleep(1)
    search_button = driver.find_element('xpath', '//*[@id="bline_sch"]/dl/dd/button')
    search_button.click()
    time.sleep(1)
    result_iframe = driver.find_element('xpath', '//*[@id="ifResult"]')
    driver.switch_to.frame(result_iframe)
    
    bline_results = driver.find_elements('xpath', '//*[@class="bline_result"]')
    print(len(bline_results))
    for bline_result in bline_results:
        span = bline_result.find_element('xpath', './/div/p/span')
        bus_number = span.text
        print(bus_number)
        
        if bus_number in bus_numbers:
            print(bus_number, ': passed')
            continue
        else:
            bus_numbers.append(bus_number)
            
        expand_button = bline_result.find_element('xpath', './/div[1]/p[2]/a[2]')
        # expand_button = bline_result.find_element('xpath', '(//div)[0]/p[@class=".btmore" ]/a[1]')
        # print('child: ',expand_button.text)
        # print(expand_button.get_attribute("href"))
        expand_button.click()
        time.sleep(3)
        station_list = bline_result.find_elements('xpath', './/div[3]/table/tbody/tr')
        for station in station_list:

            station_infos_raw = station.find_element('xpath', './/td[2]/a')
            station_infos = station_infos_raw.get_attribute("href")
            station_infos = station_infos.replace('javascript:statonARSMiddle(', '')
            station_infos = station_infos.replace(')', '')
            station_info_list = station_infos.split(',')
            latitude = station_info_list[0].strip()
            longitude = station_info_list[1].strip()
            ars_number = station.find_element('xpath', './/td[1]').text
            station_name = station.find_element('xpath', './/td[2]').text
            station_code = station_info_list[4].strip().replace('\'', '')

    #         print(latitude)
    #         print(longitude)
    #         print(ars_number)
    #         print(station_name)
    #         print(station_code)
            text = '' + bus_number + ',' + station_code + ',' + station_name + ',' + ars_number + ',' + latitude + ',' + longitude
            # print(text)
            write(text)
            # break;
        time.sleep(1)
    driver.switch_to.default_content()
```
