```python
# 버스 번호별 일자별 재차인원을 다운로드 받는다
def download_person_count_per_day(target_date, target_bus_num):
    #target_date = '2021-06-18'
    #target_num = 
    script_text = """
    var searchDay = "{0}";
    $("#searchFromDay").val(searchDay);
    $("#searchToDay").val(searchDay);
    $("#tcboId").val("03");
    $("#routeId").val("{1}")
    $("#routeSdCd").val("26");
    $("#routeSggCd").val("26");

    """
    final_text = script_text.format(target_date, target_bus_num)
    print(final_text)
    driver.execute_script(final_text);
    
    # 조회 버튼 클릭
    driver.find_element_by_xpath('//*[@id="btnSearch"]/button').click()
    time.sleep(3)
    
    # 노선 선택 경고창 표시시
    try:
        alert = driver.switch_to.alert
        if alert is not None and alert.text == '노선을 선택해 주세요.':
            return False
    except:
        print("노선 선택 경고 메시지 없음")
    
    empty_text = driver.find_elements_by_xpath('//*[@id="table_div"]/table/tbody/tr/td[contains(text(), "데이터가 없습니다")]')
    if empty_text:
        print(empty_text[0].get_attribute("innerHTML").splitlines()[0])
        driver.find_element_by_xpath('//*[@class="but_area_back"]/button').click()
        return False
    else:
        # csv 다운로드 버튼 클릭
        try:
            download_link = driver.find_element_by_xpath('//*[@id="rgrstyReportResult"]/h2/p/span/a')
        except:
            print('button error')
        
        for i in range(0, 10):
            if download_link is None:
                time.sleep(1)
                try:
                    download_link = driver.find_element_by_xpath('//*[@id="rgrstyReportResult"]/h2/p/span/a')
                except:
                    print('button error')

        if download_link is not None:
            download_link.click()
            # 팝업 YES 클릭
            try:
                WebDriverWait(driver, 3).until(EC.alert_is_present(),
                                               'Timed out waiting for PA creation ' +
                                               'confirmation popup to appear.')
                alert = drive.switch_to.alert
                alert.accept()
                print('alert accepted')
            except Exception:
                print('no alert')

            # 돌아가기 클릭
            # sleep(1)
            driver.find_element_by_xpath('//*[@class="but_area_back"]/button').click()
            return True
        else:
            driver.find_element_by_xpath('//*[@class="but_area_back"]/button').click()
            return False
```

```python
# 다운로드 받은 파일의 이름은 항상 일정하다. 버스번호, 버스코드, 해당날짜로 파일 이름을 바꿔준다
def change_file_name(target_bus_num, target_bus_code, target_date):
    
    today = date.today().strftime("%Y%m%d")
    downloaded_file_name = '노선·정류장 지표(노선별 차내 재차인원)_' + today + '.csv'
    target_file_name = "Bus_" + target_bus_num + "_" + target_bus_code + "_" + target_date + ".csv"
    downloads_path = str(Path.home() / "Downloads")
    file_path = downloads_path + "/" + downloaded_file_name
    if os.path.isfile(file_path):
        os.rename(file_path, downloads_path + "/" + target_file_name)
```
